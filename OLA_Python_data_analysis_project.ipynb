```python
# plotting a bar chart for Payment_Method and it's count

ax = sns.countplot(x = 'Payment_Method',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Vehicle_Type and Payment_Method

ax = sns.countplot(data = df, x = 'Vehicle_Type', hue = 'Payment_Method')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Booking_Status And Booking_Value

sales_state = df.groupby(['Booking_Status'], as_index=False)['Booking_Value'].sum().sort_values(by='Booking_Value', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Booking_Status',y= 'Booking_Value')
```

```python
# Vehicle_Type And Booking_Value

sales_state = df.groupby(['Vehicle_Type'], as_index=False)['Booking_Value'].sum().sort_values(by='Booking_Value', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Vehicle_Type',y= 'Booking_Value')
```

```python
# Vehicle Type vs Booking Status
# Yeh tumhe kaunse vehicle me cancellation zyada hai batayega

sns.countplot(x='Vehicle_Type', hue='Booking_Status', data=df)
plt.title("Vehicle Type vs Booking Status")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Pickup Locations

top_pickup = df['Pickup_Location'].value_counts().head(10)

top_pickup.plot(kind='bar')
plt.title("Top 10 Pickup Locations")
plt.xticks(rotation=45)
plt.show()
```

```python
# Ride Trend Over Time

df.groupby(df['Date'].dt.date).size().plot()
plt.title("Daily Ride Trend")
plt.xlabel("Date")
plt.ylabel("Number of Rides")
plt.show()
```

```python
# Cancellation Comparison

plt.figure()

customer_cancellations = (df['Canceled_Rides_by_Customer'] != 0).sum()
driver_cancellations = (df['Canceled_Rides_by_Driver'] != 0).sum()

cancel_data = pd.Series({'Canceled by Customer': customer_cancellations,
                         'Canceled by Driver': driver_cancellations})

cancel_data.plot(kind='bar')
plt.title("Cancellation Comparison")
plt.ylabel("Count")
plt.show()
```

```python
#  Heatmap (Advanced)

df['Hour'] = df['Date'].dt.hour
plt.figure()
pivot = df.pivot_table(index='Hour', columns='Booking_Status', aggfunc='size', fill_value=0)
sns.heatmap(pivot)
plt.title("Heatmap: Hour vs Booking Status")
plt.show()
```
